```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
head_position_y = np.zeros(400)
hand_position_y = np.ones(400)
head_position_y[:20] = (np.random.rand(20))
hand_position_y[15:20] = -(np.random.rand(5))
for i in np.arange(20,100):
    head_position_y[i] = 1 + math.sin(i) + 0.3 * (np.random.rand(1))
    hand_position_y[i] = head_position_y[i] * 1.1
head_position_y[100:200] = (np.random.rand(100))
hand_position_y[100:115] = head_position_y[100:115] * 0.9
hand_position_y[115:200] = head_position_y[115:200] * 1.3
head_position_y[200:220] = np.random.rand(20)
for i in np.arange(220,400):
    head_position_y[i] = 1 + math.sin(i) + 0.3 * (np.random.rand(1))
hand_position_y[200:220] = head_position_y[200:220]*np.random.rand(20)
hand_position_y[220:] = head_position_y[220:] * 1.1



# # hand_position_y[]
# for i in np.arange(300,400):
#     head_position_y[i] = math.sin(i) + 0.1 * np.random.rand(1)
# hand_position_y = np.ones(400)

# hand_position_y[20:200] = 1.1 * head_position_y[20:200]
# hand_position_y[200:220] = np.random.rand(20) * head_position_y[200:220]
# hand_position_y[200:220]
```

## example dataframe

```python
df = pd.DataFrame({'4-YRWrist': hand_position_y, '7-YLWrist' : hand_position_y, 'YHead': head_position_y})
```

```python
"""if hand is raised the interval is x.5
if we are in a rest period the interval is even (mod(interval, 2)==0)
if we are in a workout period the iterval is odd (mod(interval, 2)==1)"""
interval = 0
df['interval'] = np.zeros(df.shape[0])
row = 1
while row<df.shape[0]:
#     print(np.sign(np.min([df.loc[row,'4-YRWrist'], df.loc[row,'7-YLWrist']]) - df.loc[row, 'YHead']))
        
    if np.sign(np.min([df.loc[row,'4-YRWrist'], df.loc[row,'7-YLWrist']]) - df.loc[row, 'YHead']) != np.sign(np.min([df.loc[row - 1,'4-YRWrist'], df.loc[row - 1,'7-YLWrist']]) - df.loc[row - 1, 'YHead']):
        interval += 0.5
#         print(df.loc[row,'interval'])
        
    df.interval[row] = interval
    row += 1

number_of_stages = len(df.interval.value_counts()) #the different stages    
interval = 0
interval_frames = []
for i in np.arange(number_of_stages):
    interval_frames.append(df[df.interval==interval])
    interval += 0.5
    


    
```
